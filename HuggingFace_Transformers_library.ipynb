```python
from transformers import pipeline

# Create a text generation pipeline using GPT-2
generator = pipeline('text-generation', model='gpt2')

# Generate text
result = generator("Once upon a time", max_length=50, num_return_sequences=1)

print(result)
```

```python
from transformers import AutoTokenizer

# Load pre-trained tokenizer
tokenizer = AutoTokenizer.from_pretrained("bert-base-uncased")

# Tokenize a sample text
sample_text = "Hello, how are you?"
tokens = tokenizer.tokenize(sample_text)
token_ids = tokenizer.convert_tokens_to_ids(tokens)

print("Tokens:", tokens)
print("Token IDs:", token_ids)
```

```python
from transformers import AutoModelForSequenceClassification

# Load pre-trained model
model = AutoModelForSequenceClassification.from_pretrained("bert-base-uncased")

print(model)
```

```python
from transformers import pipeline

# Create a text classification pipeline
classifier = pipeline('sentiment-analysis')

# Classify text
result = classifier("I love using the HuggingFace library!")

print(result)
```

```python
from transformers import pipeline

# Create a NER pipeline
ner = pipeline('ner', model='dbmdz/bert-large-cased-finetuned-conll03-english')

# Recognize entities in text
result = ner("HuggingFace Inc. is a company based in New York City. Its headquarters are in DUMBO, therefore very close to the Manhattan Bridge.")

print(result)
```

```python
from transformers import pipeline

# Create a QA pipeline
qa = pipeline('question-answering')

# Answer questions
result = qa(question="Where is HuggingFace located?", context="HuggingFace Inc. is a company based in New York City.")

print(result)
```
